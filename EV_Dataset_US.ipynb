```python
# df['fuel_economy','Number of Metro Organizing Committees','affectweather','devharm','discuss','exp','localofficials','personal','reducetax','regulate','worried ','gasoline_price_per_gallon'].fillna(df['fuel_economy','Number of Metro Organizing Committees','affectweather','devharm','discuss','exp','localofficials','personal','reducetax','regulate','worried ','gasoline_price_per_gallon'].median(), inplace=True)
#  mot the correct way to write this it should be like this
# Strip any leading/trailing whitespace from column names
df.columns = df.columns.str.strip()

# Define the columns to fill
cols_to_fill = [
    'fuel_economy',
    'Number of Metro Organizing Committees',
    'affectweather',
    'devharm',
    'discuss',
    'exp',
    'localofficials',
    'personal',
    'reducetax',
    'regulate',
    'worried',
    'gasoline_price_per_gallon',
    'Total',
    'Trucks',
    'Trucks_Share'
]

# Fill missing values in these columns using their respective medians
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].median())

# Confirm all values in the specified columns are now filled
df[cols_to_fill].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Party']=df['Party'].fillna(df['Party'].mode()[0])
```

```python


sns.boxplot(x='Party', y='EV Share (%)', data=df)
plt.title('EV Share (%) by Political Party')            # Outliers means the percentage of EV in different states with same party but  % EV adoption
plt.show()
```

```python
top_states = df.groupby('state')['EV Share (%)'].mean().sort_values(ascending=False).head(10)
top_states.plot(kind='barh', color='green')
plt.xlabel('EV Share (%)')
plt.title('Top 10 States by EV Share (%)')
plt.gca().invert_yaxis()                                  # in this step we are   taking mean of the states value and plot
plt.show()
```

```python
sns.scatterplot(x='Stations', y='EV Share (%)', hue='Party', data=df)
plt.title('EV Share vs. Number of Charging Stations')
plt.show()
```

```python
sns.regplot(x='gasoline_price_per_gallon', y='EV Share (%)', data=df)
plt.title('EV Share vs. Gasoline Price per Gallon')
plt.show()
```

```python


plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr[['EV Share (%)']].sort_values(by='EV Share (%)', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation with EV Share (%)')
plt.show()
```
